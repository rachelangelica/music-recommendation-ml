```python
df['Popularity'] = df['Popularity'].apply(str)
```

```python
new_df = df[['Name','Artist','Album','Lyrics']]
new_df
```

```python
new_df['Artist'] = new_df['Artist'].apply(lambda x:x.lower())
new_df['Album'] = new_df['Album'].apply(lambda x:x.lower())
new_df['Lyrics'] = new_df['Lyrics'].apply(lambda x:x.lower())
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2000)
```

```python
artist_vectors = cv.fit_transform(new_df['Artist']).toarray()
artist_vectors.shape
```

```python
album_vectors = cv.fit_transform(new_df['Album']).toarray()
album_vectors.shape
```

```python
lyrics_vectors = cv.fit_transform(new_df['Lyrics']).toarray()
lyrics_vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
artist_similarity = cosine_similarity(artist_vectors)
album_similarity = cosine_similarity(album_vectors)
lyrics_similarity = cosine_similarity(lyrics_vectors)
```

```python
similarity = artist_similarity * 0.3 + album_similarity * 0.2 + lyrics_similarity * 0.5
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
new_df.rename(columns={'Name':'title'}, inplace = True)
```

```python
def recommend(music):
    music_index = new_df[new_df['title'] == music].index[0]
    distances = similarity[music_index]
    music_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('My Sweet Lord')
```
